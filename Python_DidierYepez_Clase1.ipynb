```python
## EJERCICIO CONVERTIR TEMPERATURAS DE FARENHEITH A CELSIUS
def EvaluarTemperatura(num):
    centigrados = (num - 32) * (5/9)
    if(centigrados < 35):
        print("Escalofrios")
    elif(centigrados > 35 and centigrados <38):
        print ("Normal")
    else:
        print ("Fiebre")
```

```python
temperaturas = [33,30,36,37,38,39,40]

list(map(EvaluarTemperatura,temperaturas))


```
